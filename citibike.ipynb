```python
# Dependencies
import pandas as pd
```

```python
# List of file names to be combined 
citifiles_2016 = ['201601-citibike-tripdata.csv', '201602-citibike-tripdata.csv', '201603-citibike-tripdata.csv'
                #'201604-citibike-tripdata.csv','201605-citibike-tripdata.csv','201606-citibike-tripdata.csv'
                 ]
                 


# Create an empty list to store each dataframe
citi_data_2016 = []

# Loop through each file in the list
for file in citifiles_2016:
    # Read the file into a dataframe
    bikedata_2016 = pd.read_csv(file)
     # Append the dataframe to the list
    citi_data_2016.append(bikedata_2016)
    
# Concatenate all dataframes in the list into one dataframe
bike_df= pd.concat(citi_data_2016, axis = 0)

# Show dataframe
bike_df.head()
```

```python
# Dictionary mapping values in the 'gender' column to new values
mapping = {0: 'Unknown', 1: 'Male', 2: 'Female'}

# Use the 'map' method to change the values in the 'gender' column
bike_df['gender'] = bike_df['gender'].map(mapping)

# Show DataFrame
bike_df.head()
```

```python
bike_df['starttime'] = pd.to_datetime(bike_df['starttime'])
bike_df.dtypes
```

```python
bike_df['stoptime'] = pd.to_datetime(bike_df['stoptime'])
bike_df.dtypes
```
